```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import numpy as np
import datetime as dt
import pandas as pd
from random import seed
from random import randint
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from scipy.signal import butter, filtfilt, hilbert
import sys
import mdptoolbox
np.set_printoptions(threshold=sys.maxsize)
```

# Data Load

```python
df_metocean = pd.read_csv("Scotland_metocean_removed.csv")
df_metocean = df_metocean.drop(['Hs (m)'], axis=1)
df_metocean['v10_pri'] = df_metocean['v10 (m/s)'].shift(1)
df_metocean['Beaufort'] = 0
bf =  np.array([0.5,1.5,3.3,5.5,7.9,10.7,13.8,17.1,20.7,40]) #0 - 9+ totla 10 scale
for i in range(1,len(bf)):
    df_metocean.Beaufort = np.where(((df_metocean['v10 (m/s)'] < bf[i]) & (df_metocean['v10 (m/s)'] > bf[i-1])),i,df_metocean.Beaufort)
df_metocean['Beaufort_pri'] = 0
df_metocean['Beaufort_pri'] = df_metocean['Beaufort'].shift(1)
df_metocean['timeindex'] = df_metocean.index%1460
```

# Season Identification Using Beaufort Scale

```python
BF = np.zeros([14599,3])
BF[:,0] = df_metocean['Beaufort_pri'][1:]
BF[:,1] = df_metocean['Beaufort'][1:]
BF[:,2] = df_metocean['timeindex'][1:]
wcss_BF = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(BF)
    wcss_BF.append(kmeans.inertia_)
plt.plot(range(1, 11), wcss_BF)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
pred_BF = kmeans.fit_predict(BF)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(BF[:,0], BF[:,1], BF[:,2], s=0.5, c=pred_BF, cmap='viridis')
ax.set_xlabel('B(t)')
ax.set_ylabel('B(t-1)')
ax.set_zlabel('Time Index')
ax.view_init(15, 10)
plt.savefig('seasons',dpi=300, bbox_inches='tight')
plt.show()

print('The season starts on ',min(np.where(kmeans.labels_[0:1459] == 0)[0])/4,'days = May 3rd')
print('The season ends   on ',max(np.where(kmeans.labels_[0:1459] == 0)[0])/4,'days = Sep 1st')
```

```python
cmf_summer = np.zeros(shape=(10,10), dtype=int)
cmf_winter = np.zeros(shape=(10,10), dtype=int) 

for i in range(1,df_metocean.shape[0]):
    if ((df_metocean['timeindex'][i] >= 489) & (df_metocean['timeindex'][i] <= 976)):
        cmf_summer[int(df_metocean['Beaufort_pri'][i]),int(df_metocean['Beaufort'][i])] += 1
    else:
        cmf_winter[int(df_metocean['Beaufort_pri'][i]),int(df_metocean['Beaufort'][i])] += 1
        
cmf_summer_n = np.transpose(np.divide(np.transpose(cmf_summer),np.transpose(np.sum(cmf_summer,axis=1))))       
cmf_winter_n = np.transpose(np.divide(np.transpose(cmf_winter),np.transpose(np.sum(cmf_winter,axis=1))))  
```

# Solving MDP -  Summer

```python
P = np.zeros((2,10,10))
P[0] = cmf_summer_n
P[1] = cmf_summer_n

c_down = -100
c_fail = -1
R = np.zeros((10,2))
R[:,0] = c_down
R[0:6,1] = 0
R[6:,1] = c_down + c_fail
```

```python
vi = mdptoolbox.mdp.ValueIteration(P,R,0.96)
vi.run()
vi.policy
```

# Solving MDP - Winter

```python
P = np.zeros((2,10,10))
P[0] = cmf_winter_n
P[1] = cmf_winter_n

c_down = -0.1
c_fail = -0.1
R = np.zeros((10,2))
R[:,0] = c_down
R[0:6,1] = 0
R[6:,1] = c_down + c_fail
```

```python
vi = mdptoolbox.mdp.ValueIteration(P,R,0.96)
vi.run()
vi.policy
```

```python
WorkTime = 4

CmfSummerWTC = np.zeros(shape=(20,20), dtype=int)
CmfWinterWTC = np.zeros(shape=(20,20), dtype=int)

# Elapsed Time Safe Weather Truncation: 19+ time steps will be considered as 19hrs 
elaps_trnc = ElapsedTimeSafeWeather.copy()
elaps_trnc[np.where(ElapsedTimeSafeWeather >= 19)] = 19

for i in range(WorkTime,len(elaps_trnc)):
    if i%1460 >= 489 and i%1460 <= 975:
        CmfSummerWTC[elaps_trnc[i-WorkTime],elaps_trnc[i]] += 1
    else:
        CmfWinterWTC[elaps_trnc[i-WorkTime],elaps_trnc[i]] += 1
        
TransitionSummerWTC = np.divide(CmfSummerWTC,np.sum(CmfSummerWTC,1)[:,None])
TransitionWinterWTC = np.divide(CmfWinterWTC,np.sum(CmfWinterWTC,1)[:,None])
```

```python
TransitionSummerWTC.shape
```

# Solving MDP-Summer

```python
P = np.zeros((2,len(TransitionSummerWTC)*20,len(TransitionSummerWTC)*20))
# Action 0 : Stay
for i in range(20):
    if (WorkTime + i) <= 19:
        P[0,20*i:20*(i+1),20*(WorkTime+i):20*(WorkTime+i+1)] = TransitionSummerWTC
    else:
        P[0,20*i:20*(i+1),20*19:20*20] = TransitionSummerWTC

#Action 1 : Go
for i in range(20):
    P[1,20*i:20*(i+1),WorkTime:20] = TransitionSummerWTC[:,WorkTime:]
    
    if (WorkTime + i) <= 19:
        P[1,20*i:20*(i+1),20*(WorkTime+i):20*(WorkTime+i)+WorkTime] = TransitionSummerWTC[:,:WorkTime]
    else:
        P[1,20*i:20*(i+1),20*19:20*19+WorkTime] = TransitionSummerWTC[:,:WorkTime]
```

```python
c_down = -1
c_fail = -10
R = np.zeros((len(TransitionSummerWTC)*20,2))

for i in range(4,20):
    R[i*20:,0] = c_down*i
    R[i*20:,1] = c_down*i
    
for i in range(20):
    R[20*i:20*i+WorkTime,1] += c_fail
```

```python
vi = mdptoolbox.mdp.ValueIteration(P,R,0.96,epsilon=0.000001)
vi.setSilent()
vi.run()
policy = np.array(vi.policy)
policy = policy.reshape((20,20))
policy
```

# Solving MDP - Winter

```python
P = np.zeros((2,len(TransitionWinterWTC)*20,len(TransitionWinterWTC)*20))
# Action 0 : Stay
for i in range(20):
    if (WorkTime + i) <= 19:
        P[0,20*i:20*(i+1),20*(WorkTime+i):20*(WorkTime+i+1)] = TransitionWinterWTC
    else:
        P[0,20*i:20*(i+1),20*19:20*20] = TransitionWinterWTC

#Action 1 : Go
for i in range(20):
    P[1,20*i:20*(i+1),WorkTime:20] = TransitionWinterWTC[:,WorkTime:]
    
    if (WorkTime + i) <= 19:
        P[1,20*i:20*(i+1),20*(WorkTime+i):20*(WorkTime+i)+WorkTime] = TransitionWinterWTC[:,:WorkTime]
    else:
        P[1,20*i:20*(i+1),20*19:20*19+WorkTime] = TransitionWinterWTC[:,:WorkTime]
```

```python
c_down = -1
c_fail = -10
R = np.zeros((len(TransitionWinterWTC)*20,2))

for i in range(4,20):
    R[i*20:,0] = c_down*i
    R[i*20:,1] = c_down*i
    
for i in range(20):
    R[20*i:20*i+WorkTime,1] += c_fail
```

```python
vi = mdptoolbox.mdp.ValueIteration(P,R,0.96,epsilon=0.000001)
vi.setSilent()
vi.run()
policy = np.array(vi.policy)
policy = policy.reshape((20,20))
policy
```

```python
df_metocean = pd.read_csv("Scotland_metocean_removed.csv")
df_metocean = df_metocean.drop(['Hs (m)'], axis=1)
df_metocean['v10_pri'] = df_metocean['v10 (m/s)'].shift(1)
df_metocean['Beaufort'] = 0
bf =  np.array([0.5,1.5,3.3,5.5,7.9,10.7,13.8,17.1,20.7,40]) #0 - 9+ totla 10 scale
for i in range(1,len(bf)):
    df_metocean.Beaufort = np.where(((df_metocean['v10 (m/s)'] < bf[i]) & (df_metocean['v10 (m/s)'] > bf[i-1])),i,df_metocean.Beaufort)
df_metocean['Beaufort_pri'] = 0
df_metocean['Beaufort_pri'] = df_metocean['Beaufort'].shift(1)
df_metocean['timeindex'] = df_metocean.index%1460
df_metocean.head()
```

```python
BeaufortSafetyThres = 5
WorkTime = 4
t_aux = 2
c_down = -1.5
c_fail = -1
cmap='viridis'
```

```python
#Elapsed time series
BinaryTimeSeries = df_metocean['Beaufort'] <= BeaufortSafetyThres
BinaryTimeSeries = BinaryTimeSeries.astype(int)
ElapsedTimeSafeWeather = np.zeros_like(BinaryTimeSeries)
if BinaryTimeSeries[0]:
    ElapsedTimeSafeWeather[0] = 1
for i in range(1,BinaryTimeSeries.shape[0]):
    if BinaryTimeSeries[i]:
        ElapsedTimeSafeWeather[i] = ElapsedTimeSafeWeather[i-1] + 1    

#Intermediate transition matrix
CmfSummerWTC = np.zeros(shape=(22,22), dtype=int)
CmfWinterWTC = np.zeros(shape=(22,22), dtype=int)

# Elapsed Time Safe Weather Truncation: 21+ time steps will be considered as 21
elaps_trnc = ElapsedTimeSafeWeather.copy()
elaps_trnc[np.where(ElapsedTimeSafeWeather > 21)] = 21

for i in range(WorkTime,len(elaps_trnc)):
    if i%1460 >= 490 and i%1460 <= 976:
        CmfSummerWTC[elaps_trnc[i-WorkTime],elaps_trnc[i]] += 1
    else:
        CmfWinterWTC[elaps_trnc[i-WorkTime],elaps_trnc[i]] += 1
        
TransitionSummerWTC = np.divide(CmfSummerWTC,np.sum(CmfSummerWTC,1)[:,None])
TransitionWinterWTC = np.divide(CmfWinterWTC,np.sum(CmfWinterWTC,1)[:,None])
```

```python
#Downtime range 0 - 9
P_summer = np.zeros((2,len(TransitionSummerWTC)*10,len(TransitionSummerWTC)*10))
P_winter = np.zeros((2,len(TransitionWinterWTC)*10,len(TransitionWinterWTC)*10))
# Action 0 : Stay
for i in range(10):
    if (WorkTime + i) <= 9:
        P_summer[0,22*i:22*(i+1),22*(WorkTime+i):22*(WorkTime+i+1)] = TransitionSummerWTC
        P_winter[0,22*i:22*(i+1),22*(WorkTime+i):22*(WorkTime+i+1)] = TransitionWinterWTC
    else:
        P_summer[0,22*i:22*(i+1),22*9:22*10] = TransitionSummerWTC
        P_winter[0,22*i:22*(i+1),22*9:22*10] = TransitionWinterWTC

#Action 1 : Go
for i in range(10):
    P_summer[1,22*i:22*(i+1),WorkTime:22] = TransitionSummerWTC[:,WorkTime:22]
    P_winter[1,22*i:22*(i+1),WorkTime:22] = TransitionWinterWTC[:,WorkTime:]
    
    if (WorkTime + i) <= 9:
        P_summer[1,22*i:22*(i+1),22*(WorkTime+i):22*(WorkTime+i)+WorkTime] = TransitionSummerWTC[:,:WorkTime]
        P_winter[1,22*i:22*(i+1),22*(WorkTime+i):22*(WorkTime+i)+WorkTime] = TransitionWinterWTC[:,:WorkTime]
    else:
        P_summer[1,22*i:22*(i+1),22*9:22*9+WorkTime] = TransitionSummerWTC[:,:WorkTime]
        P_winter[1,22*i:22*(i+1),22*9:22*9+WorkTime] = TransitionWinterWTC[:,:WorkTime]

R = np.zeros((len(TransitionSummerWTC)*10,2))

R[(t_aux+1)*22:,:]= c_down
    
for i in range(10):
    R[22*i:22*i+WorkTime,1] += c_fail

vi_summer = mdptoolbox.mdp.ValueIteration(P_summer,R,0.96)
vi_winter = mdptoolbox.mdp.ValueIteration(P_winter,R,0.96)
vi_summer.run()
vi_winter.run()
policy_summer = np.array(vi_summer.policy)
policy_winter = np.array(vi_winter.policy)
policy_summer = policy_summer.reshape((10,22))
policy_winter = policy_winter.reshape((10,22))
print(policy_summer)
print(policy_winter)
```
